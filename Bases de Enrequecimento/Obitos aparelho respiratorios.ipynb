```python
ano2014 = df[df['Ano/mês do Óbito'].str.contains("/2014", flags=re.I, regex=True,)]
ano2014.head()
```

```python
# Exibir o gráfico de linhas com marcação em cada ponto
plt.style.use("ggplot")
plt.figure(figsize = (18, 4))
sns.pointplot(x='Ano/mês do Óbito', y='SC', data=ano2014)


# Ajustar legenda
plt.xticks(rotation=0)
plt.show()
```
