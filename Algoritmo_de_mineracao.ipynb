```python
from apyori import apriori
```

```python
transacoes = []
for i in range (0, 100):
    transacoes.append([str(df.values[i, j]) for j in range(0,4)])
```

```python
regras = apriori(transacoes, min_support = 0.001, min_confidence = 0.8, min_lift = 2, min_lenght = 3)
```

```python
resultado = list(regras)
```

```python
resultado2 = [list(x) for x in resultado]
resultado2
```
